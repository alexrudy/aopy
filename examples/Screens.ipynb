# Single Screens

```python
import astropy.units as u
from aopy.atmosphere.screen import Screen
```

```python
MyScreen = Screen((30,30), 30 * u.cm, du= 56 * u.cm, L0 = 300 * u.meter)
```

```python
plt.imshow(MyScreen(),interpolation='nearest')
```
